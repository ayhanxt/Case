```python
for column in df.columns:
    unique_values = df[column].unique()
    
    # Unique değerlerin sayısını kontrol et
    if len(unique_values) > 10:
        print(f"{column}: {len(unique_values)} unique değer")
    else:
        print(f"{column}: {unique_values}")

    # Boş veya "." içeren değerleri kontrol et
    null_or_dot_values = df[df[column].isin(['', '.'])]
    if len(null_or_dot_values) > 0:
        print(f"Boş veya '.' içeren değerler ({column}):")
        print(null_or_dot_values[column])

    print()  # İşaretleme için boş bir satır ekle
```

```python
df.loc[df['Kur_bilgisi'] == 'USD', 'EAD'] *= 20.5
df.loc[df['Kur_bilgisi'] == 'USD', 'Kur_bilgisi'] = 'TRL'

df.loc[df['Kur_bilgisi'] == 'EUR', 'EAD'] *= 22.8
df.loc[df['Kur_bilgisi'] == 'EUR', 'Kur_bilgisi'] = 'TRL'
```

```python
 df[(df['Default_Flag'] == 2) | (df['Default_Flag'] == 99)]
```

```python
df = df[(df['Default_Flag'] != 2) & (df['Default_Flag'] != 99)]
```

```python
df['Product_Type'] = df['Product_Type'].replace('.', np.nan)
df['BET_Score'] = df['BET_Score'].replace('.', np.nan)
```

```python
df.head()
```

```python
df.info()
```

```python
bet_df = df[df['BET_Score'].notna()]
```

```python
bet_df.info()
```

```python
# Banka verisi için tanımlayıcı istatistikler
banka_describe = df.describe()

# BET Şirketi verisi için tanımlayıcı istatistikler
bet_describe = bet_df.describe()

# Tanımlayıcı istatistikleri birleştirme
tanimlayici_istatistikler = pd.concat([banka_describe, bet_describe], axis=1, keys=['Banka Verisi', 'BET Şirketi Verisi'])

# Tanımlayıcı istatistikleri görüntüleme
print(tanimlayici_istatistikler)
```

```python
# Ticari Müşteriler için segment bazında analizler
ticari_customers_df = df[df['Segment'].isin(['K', 'TA', 'TB', 'SB', 'SC'])]

for segment in ['K', 'TA', 'TB', 'SB', 'SC']:
    segment_df = ticari_customers_df[ticari_customers_df['Segment'] == segment]
    
    # Gözlemlenen batık oranını hesaplayalım (Banka verisi)
    batik_oran = segment_df['Default_Flag'].mean() * 100
    print(f"Segment {segment} için Gözlemlenen Batık Oranı (Banka): {batik_oran:.2f}%")
    
    # BET Şirketinin gözlemlenen batık oranını hesaplayalım (BET verisi)
    bet_segment_df = bet_df[bet_df['Segment'] == segment]
    bet_batik_oran = bet_segment_df['Default_Flag'].mean() * 100
    print(f"Segment {segment} için Gözlemlenen Batık Oranı (BET): {bet_batik_oran:.2f}%")
    
    # Skor doluluk yüzdelerini analiz edelim
    bet_score_doluluk_yuzdesi = segment_df['BET_Score'].notnull().mean() * 100
    bank_score_doluluk_yuzdesi = segment_df['Bank_Score'].notnull().mean() * 100
    print(f"Segment {segment} için BET Skor Doluluk Yüzdesi: {bet_score_doluluk_yuzdesi:.2f}%")
    print(f"Segment {segment} için Bank Skor Doluluk Yüzdesi: {bank_score_doluluk_yuzdesi:.2f}%")
    
    # BET_Score ve Banka_Score ortalamalarını hesaplayalım
    bet_score_ortalama = segment_df['BET_Score'].mean()
    bank_score_ortalama = segment_df['Bank_Score'].mean()
    print(f"Segment {segment} için BET_Score Ortalaması: {bet_score_ortalama:.2f}")
    print(f"Segment {segment} için Banka_Score Ortalaması: {bank_score_ortalama:.2f}")
    
    # EAD (risk) tutarının dağılımını inceleyelim (Banka verisi)
    plt.hist(segment_df['EAD'], bins=10)
    plt.xlabel('EAD (Risk) Tutarı')
    plt.ylabel('Frekans')
    plt.title(f'Segment {segment} için EAD (Risk) Tutarı Dağılımı (Banka)')
    plt.show()
    
    # EAD (risk) tutarının dağılımını inceleyelim (BET verisi)
    plt.hist(bet_segment_df['EAD'], bins=10)
    plt.xlabel('EAD (Risk) Tutarı')
    plt.ylabel('Frekans')
    plt.title(f'Segment {segment} için EAD (Risk) Tutarı Dağılımı (BET)')
    plt.show()
    
    print("\n")


# Bireysel Müşteriler için ürün bazında analizler
bireysel_customers_df = df[df['Segment'].isin(['BB', 'BA', 'BO'])]

product_types = bireysel_customers_df['Product_Type'].unique()

for product_type in product_types:
    product_type_df = bireysel_customers_df[bireysel_customers_df['Product_Type'] == product_type]
    
    # Gözlemlenen batık oranını hesaplayalım (Banka verisi)
    batik_oran = product_type_df['Default_Flag'].mean() * 100
    print(f"Ürün Tipi {product_type} için Gözlemlenen Batık Oranı (Banka): {batik_oran:.2f}%")
    
    # BET Şirketinin gözlemlenen batık oranını hesaplayalım (BET verisi)
    bet_product_type_df = bet_df[bet_df['Product_Type'] == product_type]
    bet_batik_oran = bet_product_type_df['Default_Flag'].mean() * 100
    print(f"Ürün Tipi {product_type} için Gözlemlenen Batık Oranı (BET): {bet_batik_oran:.2f}%")
    
    # Skor doluluk yüzdelerini analiz edelim
    bet_score_doluluk_yuzdesi = product_type_df['BET_Score'].notnull().mean() * 100
    bank_score_doluluk_yuzdesi = product_type_df['Bank_Score'].notnull().mean() * 100
    print(f"Ürün Tipi {product_type} için BET Skor Doluluk Yüzdesi: {bet_score_doluluk_yuzdesi:.2f}%")
    print(f"Ürün Tipi {product_type} için Bank Skor Doluluk Yüzdesi: {bank_score_doluluk_yuzdesi:.2f}%")
    
    # BET_Score ve Banka_Score ortalamalarını hesaplayalım
    bet_score_ortalama = product_type_df['BET_Score'].mean()
    bank_score_ortalama = product_type_df['Bank_Score'].mean()
    print(f"Ürün Tipi {product_type} için BET_Score Ortalaması: {bet_score_ortalama:.2f}")
    print(f"Ürün Tipi {product_type} için Banka_Score Ortalaması: {bank_score_ortalama:.2f}")
    
    # EAD (risk) tutarının dağılımını inceleyelim (Banka verisi)
    plt.hist(product_type_df['EAD'], bins=10)
    plt.xlabel('EAD (Risk) Tutarı')
    plt.ylabel('Frekans')
    plt.title(f'Ürün Tipi {product_type} için EAD (Risk) Tutarı Dağılımı (Banka)')
    plt.show()
    
    # EAD (risk) tutarının dağılımını inceleyelim (BET verisi)
    plt.hist(bet_product_type_df['EAD'], bins=10)
    plt.xlabel('EAD (Risk) Tutarı')
    plt.ylabel('Frekans')
    plt.title(f'Ürün Tipi {product_type} için EAD (Risk) Tutarı Dağılımı (BET)')
    plt.show()
    
    print("\n")
```
